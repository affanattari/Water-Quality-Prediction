# Oulier Removal Process, Splitting data and Standard Scaler

```python
columns_to_process = ['ph','Hardness','Solids','Chloramines','Sulfate','Conductivity','Organic_carbon','Trihalomethanes','Turbidity']

for column_name in columns_to_process:
    Q1 = df[column_name].quantile(0.25)
    Q3 = df[column_name].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    data[column_name] = data[column_name].apply(lambda x: x if (lower_bound <= x <= upper_bound) else data[column_name].median())

# Save the cleaned dataset after outlier removal
df.to_csv('wp_after_outlier_removal.csv', index=False)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split

data_after_outlier_removal = pd.read_csv('wp_after_outlier_removal.csv')

X_after_outlier = data_after_outlier_removal.drop('Potability', axis=1)
y_after_outlier = data_after_outlier_removal['Potability']

X_train_after, X_test_after, y_train_after, y_test_after = train_test_split(X_after_outlier, y_after_outlier, test_size=0.2, random_state=42)
```

```python
X_train_after.head(20)
```

```python
import warnings
warnings.filterwarnings("ignore")

from sklearn.preprocessing import StandardScaler
st_x= StandardScaler()
X_train_after = st_x.fit_transform(X_train_after)
X_test_after = st_x.transform(X_test_after)
```

# Logistic Regression with ROC-AUC and accuracy

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
```

```python
log_reg_model = LogisticRegression()
log_reg_model.fit(X_train_after, y_train_after)
y_pred_log_reg_model = log_reg_model.predict(X_test_after)
accuracy = accuracy_score(y_test_after, y_pred_log_reg_model)
print("Accuracy:", accuracy)
```

```python
log_reg_model.fit(X_train_after,y_train_after)
y_pre_log_reg_model = log_reg_model.predict(X_train_after)
y_pred_log_reg_model = log_reg_model.predict(X_test_after)
y_pred1_log_reg_model = log_reg_model.predict_proba(X_test_after)


from sklearn.metrics import accuracy_score,roc_auc_score
from sklearn.model_selection import cross_val_score
print("Training Accuracy_score: {}".format(accuracy_score(y_train_after,y_pre_log_reg_model)))
print("Testing Accuracy_score: {}".format(accuracy_score(y_test_after,y_pred_log_reg_model)))
print("roc_auc_score: {}".format(roc_auc_score(y_test_after,y_pred1_log_reg_model[:,1])))
print("CV_score: {}".format(cross_val_score(log_reg_model, X_after_outlier, y_after_outlier, cv=10, scoring='accuracy').mean()))
print(classification_report(y_test_after, y_pred_log_reg_model))
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn import metrics
from sklearn.metrics import confusion_matrix
import seaborn as sns

c_values = list(np.arange(1, 10))

param_grid = [
    {'C': c_values, 'penalty': ['l1'], 'solver' : ['liblinear'], 'multi_class' : ['ovr']},
    {'C': c_values, 'penalty': ['l2'], 'solver' : ['liblinear', 'newton-cg', 'lbfgs'], 'multi_class' : ['ovr']}
]
grid = GridSearchCV(log_reg_model, param_grid, cv=10, scoring='accuracy')
grid.fit(X_train_after,y_train_after)
print("Best paramteres :",grid.best_params_)
print("Best score :",grid.best_score_)
print("Best estimator :",grid.best_estimator_)
predictionforest = grid.best_estimator_.predict(X_test_after)
y1_pred2=grid.best_estimator_.predict_proba(X_test_after)


print(confusion_matrix(y_test_after,predictionforest))
#print ("{0}".format(metrics.confusion_matrix(y_test, y_pred, labels=[1, 0])))
print ("Classification Report")


from sklearn.metrics import classification_report, confusion_matrix

cm = confusion_matrix(y_test_after,predictionforest)
sns.heatmap(cm, annot=True, fmt='d').set_title('Confusion matrix of linear SVM')

print ("Classification Report")
print(classification_report(y_test_after, y_pred_log_reg_model))
# labels for set 1=True to upper left and 0 = False to lower right
# print ("{0}".format(metrics.classification_report(y1_test, predictionforest, labels=[1, 0])))
acc_hyper = accuracy_score(y_test_after,predictionforest)
print (acc_hyper)
print("roc_auc_score after hypertuning: {}".format(roc_auc_score(y_test_after,y1_pred2[:,1])))
print("roc_auc_score original: {}".format(roc_auc_score(y_test_after,y_pred1_log_reg_model[:,1])))
cv_log=cross_val_score(log_reg_model, X_after_outlier,y_after_outlier , cv=10, scoring='accuracy').mean();
print("CV_score: {}".format(cross_val_score(log_reg_model, X_after_outlier,y_after_outlier, cv=10, scoring='accuracy').mean()))
```

### Hard Voting Soft Voting with decision tree

```python
from sklearn.ensemble import VotingClassifier


hard_voting_model = VotingClassifier(estimators=[
    ('log_reg', log_reg_model),
    ('svm', svm_model),
    ('dt', dt_model),
    ('rf', rf_model)
], voting='hard')


hard_voting_model.fit(X_train_after, y_train_after)

# Make predictions using Hard Voting
y_hard_pred = hard_voting_model.predict(X_test_after)

# Calculate accuracy for Hard Voting
accuracy_hard_voting = accuracy_score(y_test_after, y_hard_pred)
print("Accuracy (Hard Voting):", accuracy_hard_voting)

# Create the Voting Classifier (Soft Voting)
soft_voting_model = VotingClassifier(estimators=[
    ('log_reg', log_reg_model),
    ('svm', svm_model),
    ('dt', dt_model),
    ('rf', rf_model)
], voting='soft')

# Train the Soft Voting Classifier
soft_voting_model.fit(X_train_after, y_train_after)

# Make predictions using Soft Voting
y_soft_pred = soft_voting_model.predict(X_test_after)

# Calculate accuracy for Soft Voting
accuracy_soft_voting = accuracy_score(y_test_after, y_soft_pred)
print("Accuracy (Soft Voting):", accuracy_soft_voting)
```

```python
y_pre_hard_voting_model = hard_voting_model.predict(X_train_after)
y_pred_hard_voting_model = hard_voting_model.predict(X_test_after)
y_pred_test = [clf.predict(X_test_after) for clf in hard_voting_model.estimators_]
ensemble_pred_test = np.round(np.mean(y_pred_test, axis=0))

print("Training Accuracy_score: {}".format(accuracy_score(y_train_after,y_pre_hard_voting_model)))
print("Testing Accuracy_score: {}".format(accuracy_score(y_test_after,y_pred_hard_voting_model)))
print("roc_auc_score: {}".format(roc_auc_score(y_test_after, ensemble_pred_test)))
print("CV_score: {}".format(cross_val_score(hard_voting_model, X_after_outlier, y_after_outlier, cv=10, scoring='accuracy').mean()))
print(classification_report(y_test_after, y_pred_hard_voting_model))
```

```python
soft_voting_model.fit(X_train_after,y_train_after)
y_pre_soft_voting_model = soft_voting_model.predict(X_train_after)
y_pred_soft_voting_model = soft_voting_model.predict(X_test_after)
y_pred1_soft_voting_model = soft_voting_model.predict_proba(X_test_after)

print("Training Accuracy_score: {}".format(accuracy_score(y_train_after,y_pre_soft_voting_model)))
print("Testing Accuracy_score: {}".format(accuracy_score(y_test_after,y_pred_soft_voting_model)))
print("roc_auc_score: {}".format(roc_auc_score(y_test_after,y_pred1_soft_voting_model[:,1])))
print("CV_score: {}".format(cross_val_score(soft_voting_model, X_after_outlier, y_after_outlier, cv=10, scoring='accuracy').mean()))
print(classification_report(y_test_after, y_pred_soft_voting_model))
```

```python
import warnings
from sklearn.metrics import accuracy_score, roc_auc_score, roc_curve
import matplotlib.pyplot as plt

roc_auc_log_reg = roc_auc_score(y_test_after,y_pred1_log_reg_model[:,1])
roc_auc_svm = roc_auc_score(y_test_after,y_pred1_svm_model[:,1])
roc_auc_decision_tree = roc_auc_score(y_test_after,y_pred1_dt_model[:,1])
roc_auc_rf = roc_auc_score(y_test_after,y_pred1_rf_model[:,1])
roc_auc_hard_voting = roc_auc_score(y_test_after, ensemble_pred_test)
roc_auc_soft_voting = roc_auc_score(y_test_after,y_pred1_soft_voting_model[:,1])
```

```python
fpr_log_reg, tpr_log_reg, _ = roc_curve(y_test_after, y_pred1_log_reg_model[:,1])
fpr_svm, tpr_svm, _ = roc_curve(y_test_after, y_pred1_svm_model[:,1])
fpr_decision_tree, tpr_decision_tree, _ = roc_curve(y_test_after, y_pred1_dt_model[:,1])
fpr_rf, tpr_rf, _ = roc_curve(y_test_after, y_pred1_rf_model[:,1])
fpr_hard_voting, tpr_hard_voting, _ = roc_curve(y_test_after, ensemble_pred_test)
fpr_soft_voting, tpr_soft_voting, _ = roc_curve(y_test_after, y_pred1_soft_voting_model[:,1])

plt.figure(figsize=(10, 6))
plt.plot(fpr_log_reg, tpr_log_reg, label=f'Logistic Regression (AUC = {roc_auc_log_reg:.3f})')
plt.plot(fpr_svm, tpr_svm, label=f'SVM (AUC = {roc_auc_svm:.3f})')
plt.plot(fpr_decision_tree, tpr_decision_tree, label=f'DT (AUC = {roc_auc_decision_tree:.3f})')
plt.plot(fpr_rf, tpr_rf, label=f'Random Forest (AUC = {roc_auc_rf:.3f})')
plt.plot(fpr_hard_voting, tpr_hard_voting, label=f'Hard Voting (AUC = {roc_auc_hard_voting:.3f})')
plt.plot(fpr_soft_voting, tpr_soft_voting, label=f'Soft Voting (AUC = {roc_auc_soft_voting:.3f})')
plt.plot([0, 1], [0, 1], 'k--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC AUC Curve')
plt.legend(loc="lower right")
plt.show()
```
